```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

file = r'ceo.xls'
df = pd.read_excel(file)
salarySeries = df['salary']
```

```python
print('1)')
print('Mean: ' + str(salarySeries.mean()))
print('Trimmed 10% mean: ' + str(stats.trim_mean(salarySeries, .1)))
print('')
print('Low 10% quantile: ' + str(salarySeries.quantile(.1)))
print('Low quartile: ' + str(salarySeries.quantile(.25)))
print('Median quartile: ' + str(salarySeries.quantile()))
print('Upper quartile: ' + str(salarySeries.quantile(.75)))
print('Upper 10% quantile: ' + str(salarySeries.quantile(.9)))
print('TODO Describe results')
```

```python
print('2)')
salarySeries.hist(cumulative=True, bins=9, normed=True)

plt.show()
```

```python
salarySeries.plot(kind='box', showfliers=False)
```

```python
print('2.1')

corr = df.corr()['salary']

sns.heatmap(corr, annot=True)
```

```python
print('2.2)')
sns.pairplot(df);
```

```python
spearmanCorrelation = df.corr(method="spearman")

spearmanCorrelation["salary"]
sns.heatmap(spearmanCorrelation, annot=True)
```
